```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
N = 20
rv = scipy.stats.weibull_min(c=5.0, scale=2.5)
T = 1.2
#T = 1e6
data = rv.rvs(size=N)
censored = data >= T
data[censored] = T

data_obs  = data[np.logical_not(censored)]
data_cens = data[censored]

px = np.linspace(0, 5, num=100)
fig = plt.figure(figsize=(12, 4))
ax = plt.subplot(121)
ax.plot(px, rv.pdf(px))
# ax = plt.subplot(122)
# ax.plot(data, np.zeros(len(data)), '|');

print data_obs
print rv.ppf(0.05)
```

# Likelihood

$$
\begin{aligned}
D_i &= (y_i, c_i) \\
P(D\:\vert\:\theta) &= \prod_{i:\,c_i=0} P(Y = y_i\:\vert\:\theta) \times \prod_{i:\,c_i=1} P(Y \geq y_i\:\vert\:\theta) \\
\end{aligned}
$$

```python
import pystan
model_code = """
data {
    int<lower=0> N_obs;
    real y_obs[N_obs];

    int<lower=0> N_cens;
    real y_cens[N_cens];
    
    real alpha_mu;
    real alpha_sigma;
    
    real sigma_mu;
    real sigma_sigma;
}

parameters {
    real<lower=0> alpha;
    real<lower=0> sigma;
}

model {
    alpha ~ normal(alpha_mu, alpha_sigma);
    sigma ~ lognormal(sigma_mu, sigma_sigma);

    y_obs ~ weibull(alpha, sigma);

    for(n in 1:N_cens)
        increment_log_prob(weibull_ccdf_log(y_cens[n], alpha, sigma));
}
"""
model = pystan.StanModel(model_code=model_code)
```

```python
fit = model.sampling(data=dict(alpha_mu = 2, alpha_sigma = .25, sigma_mu = 1.8, sigma_sigma = .5, \
                               N_obs=len(data_obs), y_obs=data_obs, N_cens=len(data_cens), y_cens=data_cens))
```

```python
print fit
```

```python
fit.plot();
```

```python
samples = fit.extract()
```

```python
alpha = samples['alpha'][:, np.newaxis]
sigma = samples['sigma'][:, np.newaxis]

xx = np.linspace(0, 12, num=100)
yy = scipy.stats.weibull_min.pdf(xx, c=alpha, scale=sigma)
```

```python
ymed = np.mean(yy, axis=0)
ymin = np.percentile(yy,  2.5, axis=0)
ymax = np.percentile(yy, 97.5, axis=0)

plt.xlabel('lifetime')
plt.ylabel('probability')
plt.fill_between(xx, ymin, ymax, alpha=0.2)
plt.plot(xx, ymed);
```
